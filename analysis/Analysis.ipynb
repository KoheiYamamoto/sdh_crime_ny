# Temporal: weekday + weekend/holiday

```python
def get_weekno(_Date):
    return 7 if datetime.strptime(_Date, '%Y-%m-%d') in us_holidays else datetime.strptime(_Date, '%Y-%m-%d').weekday()

crime_property['weekno'] = crime_property['Date'].apply(get_weekno)
daykind_volume = pd.DataFrame(crime_property.groupby(['weekno', 'Volume']).size()).reset_index()
daykind_volume['sum'] = daykind_volume['Volume'] * daykind_volume[0]
x = daykind_volume.weekno.unique().tolist()
y = [sum(daykind_volume.query('weekno == @i')['sum'].tolist()) for i in x]

%matplotlib notebook
plt.bar(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sut', 'Sun', 'Holiday'], y)
plt.title('Days of week  - Total no of property crimes')
plt.show()
plt.savefig('res/4_daykind_propertycrimes.png')
```

```python
crime_property = pd.read_csv('1property_near_sorted_count.csv')
crime_property = crime_property[crime_property['Date'] != '(blank)']
crime_personal = pd.read_csv('1personal_near_sorted_count.csv')
crime_both = pd.concat([crime_property, crime_personal])

crime_property_census = pd.DataFrame(crime_property.groupby(['Census_Tracts', 'Volume']).size()).reset_index()
crime_property_census['sum'] = crime_property_census['Volume'] * crime_property_census[0]
crime_property_census = crime_property_census.groupby('Census_Tracts').apply(lambda d: d.sum()).drop(['Census_Tracts', 'Volume', 0],axis=1).reset_index()
crime_personal_census = pd.DataFrame(crime_personal.groupby(['Census_Tracts', 'Volume']).size()).reset_index()
crime_personal_census['sum'] = crime_personal_census['Volume'] * crime_personal_census[0]
crime_personal_census = crime_personal_census.groupby('Census_Tracts').apply(lambda d: d.sum()).drop(['Census_Tracts', 'Volume', 0],axis=1).reset_index()
crime_both_census = pd.DataFrame(crime_both.groupby(['Census_Tracts', 'Volume']).size()).reset_index()
crime_both_census['sum'] = crime_both_census['Volume'] * crime_both_census[0]
crime_both_census = crime_both_census.groupby('Census_Tracts').apply(lambda d: d.sum()).drop(['Census_Tracts', 'Volume', 0],axis=1).reset_index()

tree = pd.read_csv('7tree_num.csv')
house = pd.read_csv('6housing_density.csv')
population_income = pd.read_csv('45population_income.csv')

crime_property_census = pd.merge(crime_property_census, tree, left_on='Census_Tracts', right_on='CTID', how='left') 
crime_property_census = pd.merge(crime_property_census, house, left_on='Census_Tracts', right_on='ctid', how='left') 
crime_property_census = pd.merge(crime_property_census, population_income, left_on='Census_Tracts', right_on='CT_ID', how='left') 
crime_personal_census = pd.merge(crime_personal_census, tree, left_on='Census_Tracts', right_on='CTID', how='left') 
crime_personal_census = pd.merge(crime_personal_census, house, left_on='Census_Tracts', right_on='ctid', how='left') 
crime_personal_census = pd.merge(crime_personal_census, population_income, left_on='Census_Tracts', right_on='CT_ID', how='left') 
crime_both_census = pd.merge(crime_both_census, tree, left_on='Census_Tracts', right_on='CTID', how='left') 
crime_both_census = pd.merge(crime_both_census, house, left_on='Census_Tracts', right_on='ctid', how='left') 
crime_both_census = pd.merge(crime_both_census, population_income, left_on='Census_Tracts', right_on='CT_ID', how='left') 

dropField = ['CTID', 'ctid', 'CT_ID', 'shape_area', 'building_a', 'Census_Tracts']
crime_property_census.drop(dropField, inplace=True, axis=1)
crime_property_census.dropna(inplace=True)
crime_personal_census.drop(dropField, inplace=True, axis=1)
crime_personal_census.dropna(inplace=True)
crime_both_census.drop(dropField, inplace=True, axis=1)
crime_both_census.dropna(inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score 
data_test_all = crime_property_census.sample(n=100)
data_train_all = crime_property_census.loc[list(set(crime_property_census.index.tolist())-set(data_test_all.index.tolist()))]

y_train = data_train_all['sum'].values.tolist()
data_train = data_train_all.drop("sum", 1)
x_train = data_train.values.tolist()

y_test = data_test_all['sum'].values.tolist()
data_test = data_test_all.drop("sum", 1)
x_test = data_test.values.tolist()

rfr = RandomForestRegressor(n_estimators=1000)
rfr.fit(x_train, y_train)
predict_y = rfr.predict(x_test)
r2_score = r2_score(y_test, predict_y)
print('oefficient of determination: ', r2_score)
feature = rfr.feature_importances_
label = data_train.columns[0:]
indices = np.argsort(feature)[::-1]
for i in range(len(feature)):
    print(str(i + 1) + "   " +
          str(label[indices[i]]) + "   " + str(feature[indices[i]]))

%matplotlib notebook
plt.subplot(122, facecolor='white')
plt.title('Importance of variables to property crimes')
plt.bar(
    range(
        len(feature)),
    feature[indices],
    color='blue',
    align='center')
# plt.xticks(range(len(feature)), label[indices], rotation=45)
plt.xticks(range(len(feature)),['population', 'income', 'tree', 'housing density'], rotation=45)
plt.xlim([-1, len(feature)])
plt.tight_layout()
plt.show()
plt.savefig('res/19_property_RF.png')
```

```python
def groupby_taxiz_timep(_tmp):
    _out = pd.DataFrame(index=[], columns=['Taxi_Zone', 'Time_Period', 'total'])
    for i in _tmp['Taxi_Zone'].unique().tolist():
        for j in _tmp['Time_Period'].unique().tolist():
            __tmp = _tmp.query('Taxi_Zone == @i and Time_Period == @j')
            res = pd.Series([str(int(i)), j, sum(__tmp['total'].tolist())], index=['Taxi_Zone', 'Time_Period', 'total'])
            _out = pd.concat([_out, pd.DataFrame([res])])
    return _out

def groupby_crimes_timep(_tmp):
    _out = pd.DataFrame(index=[], columns=['Census_Tracts', 'Time_Period', 'total'])
    for i in _tmp['Census_Tracts'].unique().tolist():
        for j in _tmp['Time_Period'].unique().tolist():
            __tmp = _tmp.query('Census_Tracts == @i and Time_Period == @j')
            res = pd.Series([str(int(i)), j, sum(__tmp['total'].tolist())], index=['Census_Tracts', 'Time_Period', 'total'])
            _out = pd.concat([_out, pd.DataFrame([res])])
    return _out

crime_property = pd.read_csv('1property_near_sorted_count.csv')
crime_property = crime_property[crime_property['Date'] != '(blank)']
crime_personal = pd.read_csv('1personal_near_sorted_count.csv')
crime_both = pd.concat([crime_property, crime_personal])

taxi = pd.read_csv('yellow_tripdata_2017_out.csv')

taxidata = taxi.groupby(['Taxi_Zone', 'Time_Period', 'Volume']).size().reset_index()
taxidata['total'] = taxidata['Volume'] * taxidata[0]
taxidata = groupby_taxiz_timep(taxidata)
taxidata.to_csv('res/taxi/taxi_timeperiod_dist.csv', index=False)

crime_property = crime_property.groupby(['Census_Tracts', 'Time_Period', 'Volume']).size().reset_index()
crime_property['total'] = crime_property['Volume'] * crime_property[0]
crime_property = groupby_crimes_timep(crime_property)
crime_property.to_csv('res/crime_dist/crime_property_timeperiod_dist.csv', index=False)

crime_personal = crime_personal.groupby(['Census_Tracts', 'Time_Period', 'Volume']).size().reset_index()
crime_personal['total'] = crime_personal['Volume'] * crime_personal[0]
crime_personal = groupby_crimes_timep(crime_personal)
crime_personal.to_csv('res/crime_dist/crime_personal_timeperiod_dist.csv', index=False)

crime_both = crime_both.groupby(['Census_Tracts', 'Time_Period', 'Volume']).size().reset_index()
crime_both['total'] = crime_both['Volume'] * crime_both[0]
crime_both = groupby_crimes_timep(crime_both)
crime_both.to_csv('res/crime_dist/crime_both_timeperiod_dist.csv', index=False)
```

```python
def unixAdd_frisk(_ds): 
    return int(datetime.strptime(_ds.Date + "/" + _ds.Time, '%m/%d/%Y/%H:%M:%S').timestamp())

def cal_mean_frisk(_tmp):
    _out = pd.DataFrame(index=[], columns=['Census_Tracts', 'mean'])
    for i in _tmp['boro_ct2010'].unique().tolist():
        __tmp = _tmp.query('boro_ct2010 == @i')
        res = pd.Series([str(int(i)), sum(__tmp['total'].tolist())/sum(__tmp[0].tolist())], index=['Census_Tracts', 'mean'])
        _out = pd.concat([_out, pd.DataFrame([res])])
    return _out

pastfrisk = pd.read_csv('crimes_pastfrisk_counts.csv')
pastfrisk['unix'] = pastfrisk.apply(unixAdd_frisk,axis=1)

pastfrisk_property = pastfrisk.query('Type == "PropertyCrime"')
pastfrisk_personal = pastfrisk.query('Type == "PersonalCrime"')

ct_property_frisk = pd.DataFrame(pastfrisk_property.groupby(['boro_ct2010', 'count_ct']).size()).reset_index()
ct_property_frisk['total'] = ct_property_frisk['count_ct'] * ct_property_frisk[0]
ct_property_frisk = cal_mean_frisk(ct_property_frisk)
ct_property_frisk.to_csv('res/pastfrisk/ct_property_frisk.csv', index=False)

nbr_property_frisk = pd.DataFrame(pastfrisk_property.groupby(['boro_ct2010', 'count_neighbours']).size()).reset_index()
nbr_property_frisk['total'] = nbr_property_frisk['count_neighbours'] * nbr_property_frisk[0]
nbr_property_frisk = cal_mean_frisk(nbr_property_frisk)
nbr_property_frisk.to_csv('res/pastfrisk/nbr_property_frisk.csv', index=False)

# both_property_frisk = pd.DataFrame(pastfrisk_property.groupby(['boro_ct2010', 'count_total']).size()).reset_index()
# both_property_frisk['total'] = both_property_frisk['count_total'] * both_property_frisk[0]
# cal_mean_frisk(both_property_frisk).to_csv('res/pastfrisk/both_property_frisk.csv', index=False)

ct_personal_frisk = pd.DataFrame(pastfrisk_personal.groupby(['boro_ct2010', 'count_ct']).size()).reset_index()
ct_personal_frisk['total'] = ct_personal_frisk['count_ct'] * ct_personal_frisk[0]
ct_personal_frisk = cal_mean_frisk(ct_personal_frisk)
ct_personal_frisk.to_csv('res/pastfrisk/ct_personal_frisk.csv', index=False)

nbr_personal_frisk = pd.DataFrame(pastfrisk_personal.groupby(['boro_ct2010', 'count_neighbours']).size()).reset_index()
nbr_personal_frisk['total'] = nbr_personal_frisk['count_neighbours'] * nbr_personal_frisk[0]
nbr_personal_frisk = cal_mean_frisk(nbr_personal_frisk)
nbr_personal_frisk.to_csv('res/pastfrisk/nbr_personal_frisk.csv', index=False)

# both_personal_frisk = pd.DataFrame(pastfrisk_personal.groupby(['boro_ct2010', 'count_total']).size()).reset_index()
# both_personal_frisk['total'] = both_personal_frisk['count_total'] * both_personal_frisk[0]
# cal_mean_frisk(both_personal_frisk).to_csv('res/pastfrisk/both_personal_frisk.csv', index=False)

ct_both_frisk = pd.DataFrame(pastfrisk.groupby(['boro_ct2010', 'count_ct']).size()).reset_index()
ct_both_frisk['total'] = ct_both_frisk['count_ct'] * ct_both_frisk[0]
ct_both_frisk = cal_mean_frisk(ct_both_frisk)
ct_both_frisk.to_csv('res/pastfrisk/ct_both_frisk.csv', index=False)

nbr_both_frisk = pd.DataFrame(pastfrisk.groupby(['boro_ct2010', 'count_neighbours']).size()).reset_index()
nbr_both_frisk['total'] = nbr_both_frisk['count_neighbours'] * nbr_both_frisk[0]
nbr_both_frisk = cal_mean_frisk(nbr_both_frisk)
nbr_both_frisk.to_csv('res/pastfrisk/nbr_both_frisk.csv', index=False)

# both_both_frisk = pd.DataFrame(pastfrisk.groupby(['boro_ct2010', 'count_total']).size()).reset_index()
# both_both_frisk['total'] = both_both_frisk['count_total'] * both_both_frisk[0]
# cal_mean_frisk(both_both_frisk).to_csv('res/pastfrisk/both_both_frisk.csv', index=False)
```
